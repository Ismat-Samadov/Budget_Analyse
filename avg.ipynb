```python
data =pd.read_csv('budjet_06_07_2023.csv')
data=data.sort_values(['date'],ascending=True)
data
```

```python
data['date'] = pd.to_datetime(data['date'])
current_datetime = datetime.now(timezone('UTC'))
data['time_diff'] =(current_datetime - data['date']).dt.days
data['hours']=data['date'].dt.hour
data['Weekday'] = data['date'].dt.weekday+1
data['Year'] = data['date'].dt.year
data['Day_of_Year'] = data['date'].dt.dayofyear
data['Month'] = data['date'].dt.month
data
```

```python
# Group the DataFrame by 'date' and calculate the average expenses for each day
daily_average = data[data['category'] == 'Coffe']['expenses'].sum()

print(daily_average/365)
```
